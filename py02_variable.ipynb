# Ex 1. 조건문


* 3개의 정수를 입력받고 각각 변수에 저장.
* 3개 정수의 평균을 소수점까지 계산하고 출력.
* 평균이 90점 이상이면 A, 80 이상이면 B, 70점 이상이면 C, 70점 미만이면 F를 출력.


```python
a = int(input('정수 입력>>> '))
b = int(input('정수 입력>>> '))
c = int(input('정수 입력>>> '))
sum = ((a+b+c)/3)
print('3과목 평균 = ', sum)
if sum >= 90:
    print('학점 : A')
elif 80 <= sum < 90:
    print('학점 : B')
elif 70 <= sum < 80:
    print('학점 : C')
else :
    print('학점 : F')
```
