# for 반복문

```
for 변수 in iterable_type:
    반복해서 실행할 코드 블록
```

* iterable_type: 반복가능한 타입의 객체
    * range, list, tuple, dict, set, str, ...
* `range` 함수
    * `range(stop)`:
        * 0 <= x < stop 범위의 1씩 증가하는 정수들을 생성.
    * `range(start, stop)`:
        * start <= x < stop 범위의 1씩 증가하는 정수들을 생성.
    * `range(start, stop, step)`:
        * step > 0 경우: start <= x < stop 범위의 step 만큼씩 증가하는 정수들을 생성.
        * step < 0 경우: start >= x < stop 범위의 step 만큼씩 감소하는 정수들을 생성.

````python
for x in range(5): # 0 <= x < 5
    print(x, end=' ')
````

````python
for x in range(1, 6): # 1 <= x < 6
    print(x, end=' ')
````

````python
for x in range(1, 10, 2):
    print(x, end=' ')
````

````python
for x in range(10, 1, -2):
    print (x, end=' ')
````

# 중첩 반복문

````python
# 구구단 2단부터 9단까지 출력
for x in range(2, 10):
    for y in range(1, 10):
        print(f'{x} x {y} = {x * y}')
    print('-' * 20)
````

## break와 continue

*`break`:  iteration을 종료하고 break가 포함된 가장 가까운 반복문을 빠져 나옴.

*`continue`: 반복문 내부의 실행을 멈추고, 그 다음 interation을 계속함.

````python
for x in range(5):
    if x == 2:
        continue
    print(x)
````

````python
for x in range(2, 10):
    for y in range(1, 10):
        if x < y:
            break
        print(f'{x} x {y} = {x * y}')
    print('-' * 20)
````

# while 반복문
```
[초기화 문장]
while 조건식:
    조건식 참일 때 실행할 코드 블록.
    [조건을 변경 할 수 있는 문장]
```

````python
x = ''
while x != '*****':
    x = x + '*'
    print(x)

print('-' * 20)

n = 1
while n <= 5:
    print('*' * n)
    n += 1
````

````python
# 1 이상 10 미만의 홀수들을 한줄에 출력

x = 1
while x < 10:
    print(x, end=' ')
    x += 2
````
